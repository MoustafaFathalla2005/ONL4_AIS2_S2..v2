```python
import random

def guess_game():
    num = random.randint(1, 10)  # generate random number between 1 and 10
    
    while True:
        try:
            guess = int(input('Enter a number between 1 and 10: '))
            if guess == num:
                print('Correct!')
                break
            else:
                print('Try again')
        except ValueError:
            print("Please enter a valid number.")

guess_game()
```
