### task1

```python
class Task1:
    """
    Task1 class handles Fibonacci series for n numbers.

    Attributes:
        number (int): The number of items in the Fibonacci series.
    """

    def __init__(self, number: int):
        """
        Initializes Task1 with a number.

        Args:
            number (int): The number of items from the Fibonacci series.
        """
        self.number = number

    def fibonacci(self) -> list:
        """
        Generates a Fibonacci series for n numbers.

        Returns:
            list: A list containing the Fibonacci series with n numbers.
        """
        if self.number <= 0:
            return []
        elif self.number == 1:
            return [0]
        elif self.number == 2:
            return [0, 1]

        list_fibo = [0, 1]
        for _ in range(self.number - 2):
            m = list_fibo[-1] + list_fibo[-2]
            list_fibo.append(m)
        return list_fibo



obj = Task1(10)
print(obj.fibonacci())  
```

### task2

```python
class Task2:
    """
    Task2 class handles removing a specific item from a set by its position.

    Attributes:
        s (set): The set we want to remove from.
        position (int): The index of the item to remove.
    """

    def __init__(self, s: set, position: int):
        """
        Initializes Task2 with a set and a position.

        Args:
            s (set): The set we want to remove from.
            position (int): The index of the item to remove.
        """
        self.s = s
        self.position = position

    def remove_setitem(self) -> set:
        """
        Removes a specific item from the set by its position.

        Returns:
            set: A new set without the specific item.
        """
        list_s = list(self.s)
        if self.position < 0 or self.position >= len(list_s):
            raise IndexError("Position out of range")
        list_s.pop(self.position)
        return set(list_s)



set_bremove = {1, 2, 3, 4, 5}
obj = Task2(set_bremove, 2)
print(obj.remove_setitem())
```

### task3

```python
class Task3:
    """
    Task3 class handles finding the index of an item in a tuple.

    Attributes:
        T (tuple): The tuple we want to search in.
        item (any): The item we want to find its index.
    """

    def __init__(self, T: tuple, item):
        """
        Initializes Task3 with a tuple and an item.

        Args:
            T (tuple): The tuple we want to search in.
            item (any): The item we want to find its index.
        """
        self.T = T
        self.item = item

    def index_tuple(self) -> int:
        """
        Finds the index of the item in the tuple.

        Returns:
            int: The index of the item in the tuple.
        """
        return self.T.index(self.item)


my_tuple = ('apple', 'banana', 'orange', 'grape', 'banana')
obj = Task3(my_tuple, 'orange')
print(f"Index of '{obj.item}' is: {obj.index_tuple()}")
```

```python
class Task5:
    """
    Task5 class handles if word is balindrom or not .

    Attributes:
        string (str): The string to be checked if palindrom or not.
    """

    def __init__(self, string: str):
        """
        Initializes Task5 with a string.

        Args:
            string (str): The string to be The string to be checked if palindrom or not.
        """
        self.string = string

    def is_palindrome (self)->bool:
        """
        Check if word is balindrom or not 
        
        Returns: if is palindrom or not.
        """
        return self.string == self.string[::-1]



obj = Task5("racecar")
print(obj.is_palindrome())  
```

### task7

```python
class Task7:
    """
    Task7 class handles calculating the sum of even numbers in a list.

    Attributes:
        numbers (list): The list of numbers to process.
    """

    def __init__(self, numbers: list):
        """
        Initializes Task7 with a list of numbers.

        Args:
            numbers (list): The list of numbers to process.
        """
        self.numbers = numbers

    def calc_even(self) -> int:
        """
        Calculates the sum of even numbers in the list.

        Returns:
            int: The sum of even numbers.
        """
        total = 0
        for i in self.numbers:
            if i % 2 == 0:
                total += i
        return total


obj = Task7([1, 2, 3, 4, 5, 6])
print(obj.calc_even())  
```

```python
class Task8:
    """
    Task8 class handles reversing a list.

    Attributes:
        numbers (list): The list to be reversed.
    """

    def __init__(self, numbers: list):
        """
        Initializes Task8 with a list of numbers.

        Args:
            numbers (list): The list to be reversed.
        """
        self.numbers = numbers

    def reverse_list(self) -> list:
        """
        Reverses the list.

        Returns:
            list: The reversed list.
        """
        return self.numbers[::-1]


obj = Task8([1, 2, 3])
print(obj.reverse_list())  
```

### task9

```python
def median(lst: list) -> float:
    """
    This function extracts the median from a list of numbers.

    Args:
        lst (list): List of numbers.

    Returns:
        float: The median of the list.
    """
    lst.sort()
    n = len(lst)
    mid = n // 2

    if n % 2 == 0:
        
        return (lst[mid - 1] + lst[mid]) / 2
    else:
        
        return lst[mid]



print(median([5, 2, 7, 3, 1]))      
```

### task10

```python
class Task10:
    """
    Task10 class handles finding the mode (most frequent element) in a list.

    Attributes:
        numbers (list): The list of numbers to process.
    """

    def __init__(self, numbers: list):
        """
        Initializes Task10 with a list of numbers.

        Args:
            numbers (list): The list of numbers to process.
        """
        self.numbers = numbers

    def calculate_mode(self) -> int:
        """
        Calculates the mode of the list.

        Returns:
            int: The most frequent element in the list.
        """
        counts = {num: self.numbers.count(num) for num in set(self.numbers)}
        return max(counts, key=counts.get)



obj = Task10([5, 2, 7, 3, 2, 1, 5])
print(obj.calculate_mode())  
```

```python
class Task11:
    """
    Task11 class handles converting a numeric mark to a grade.

    Attributes:
        mark (int): The student's numeric mark.
    """

    def __init__(self, mark: int):
        """
        Initializes Task11 with a student's mark.

        Args:
            mark (int): The student's numeric mark.
        """
        self.mark = mark

    def get_grade(self) -> str:
        """
        Converts the numeric mark to a grade.

        Returns:
            str: The grade (A, B, C, D, F).
        """
        if 85 <= self.mark <= 100:
            return "A"
        elif 75 <= self.mark < 85:
            return "B"
        elif 65 <= self.mark < 75:
            return "C"
        elif 50 <= self.mark < 65:
            return "D"
        else:
            return "F"

mark = int(input("Enter your mark: "))
obj = Task11(mark)
print("Your grade is:", obj.get_grade())
```
