```python
# ----
'''
Extract the core part of the message: 
"mocleW EPGTQ" from Encoded Message:(###!!@mocleW EPGTQ!!!6789).
Reverse the first word: "mocleW"becomes "Welcome".
Replace shifted vowels in the second word:o"EPGTQ":
No vowels to change.
Final decoded message: "Welcome PGTQ"
'''
Encoded_Message='###!!@mocleW EPGTQ!!!6789'
extracted_encoded_message=Encoded_Message[6:18]
first_part = extracted_encoded_message[:6][::-1]
second_part = extracted_encoded_message[7:]
filtered = []
for i in second_part:
    if i not in ['A','O','U','I','E']:
        filtered.append(i)
# memorise the join params ,pls.
second_part="".join(filtered)
decode_mesage=first_part+" "+second_part
print(decode_mesage)
```

```python
# -----
'''
Extract the core part of the message: 
"gnirtS PLIO" from Encoded Message:(&&&**$gnirtS PLIO!!@1234).
Reverse the first word: "gnirtS"becomes "String".
Replace shifted vowels in the second word:o"PLIO": 
Replace I->Eand O->Uto get "PLEU".
Final decoded message: "String PLEU".
'''
Encoded_Message='&&&**$gnirtS PLIO!!@1234'
extracted_encoded_message=Encoded_Message[6:17]
first_part = extracted_encoded_message[:6][::-1]
second_part = extracted_encoded_message[7:]
filtered = []
for i in second_part:
    if i =='I':
        i="E"
    if i =='O':
        i="U"
    filtered.append(i)
# memorise the join params ,pls.
second_part="".join(filtered)
decode_mesage=first_part+" "+second_part
print(decode_mesage)
```
