```python
def user_info(name : str , age : int):
    
    """
    this function take name and edge from user and return again
    
    :param_1 : take the name of user from input user interface
    type param_1 : str
    :param_2 : take the age of user from input user interface
    type param_2 : int
    
    return : this function will print the name and the age 
    rtype : str , int
    
    
    """
    
    return f"my name is {name} ,and my age is {age}"
```

```python
print(user_info('DEPI' , 4)) # requirement positional_calling arguments
print(user_info(age=4 , name='DEPI')) # key_word calling arguments
# you can make a func with all default args...default and need for all args in func----------------------->see this in call and in func
"".format()#see *args , and **kwargs
# if use * you can store forever variables and return this in tuple (any data type) 
# if use ** can pass a dict to function
# 5 type want to revision 
```
