```python
# factorial .............recursion
# 5!=5*4*3*2*1
# facorial(num)=num*factorial(num-1)
def factorial (num):
    """_summary_
    """
    if num == 0 :
        return
    if num == 1:
        return 1
    return  num*factorial(num-1)
```

```python
factorial(5)
```

```python
def is_prime (num:int):
    """_summary_

    Args:
        num (int): _description_
    """
    for i in range (2,num):
        if num%i == 0:
            return False
    return True
```

```python
is_prime(8)
```

```python
def reverse_txt(t):
    """_summary_

    Args:
        t (_type_): _description_
    """
    # copy above code...
    split_txt=t.split(" ")
    new_list=split_txt[::-1]
    result=" ".join(new_list)
    return result
reverse_txt('mou stafa')
```

```python
from abc import ABC , abstractmethod


class Animal(ABC):
    
    @abstractmethod
    def move(self):
        pass
    
class Bird(Animal):
    
    
    def move(self):
        print('move from bird')
        
class Cat(Animal):
    

    def move(self):
        print('move from Cat')       
        
```
